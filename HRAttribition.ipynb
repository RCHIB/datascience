```python
#creating  dataframe with only numeric variables
df_hr_num = df_hr[['Age','DailyRate','DistanceFromHome', 'Education','EmployeeNumber', 
                       'EnvironmentSatisfaction', 'HourlyRate',                     
                       'JobInvolvement', 'JobLevel','JobSatisfaction','MonthlyIncome', 
                        'MonthlyRate', 'NumCompaniesWorked','PercentSalaryHike', 'PerformanceRating', 
                       'RelationshipSatisfaction', 'StockOptionLevel',
                       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
                       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
                       'YearsWithCurrManager']].copy()
```

```python
#creating heat map for seeing relationship (strong/weak)
%matplotlib inline

plt.figure(figsize= (10,10), dpi=100)
sns.heatmap(df_hr_num.corr())
```

```python
#remove strongly correlated variables
df_hr_data = df_hr_num[['Age','DailyRate','DistanceFromHome', 
                       'EnvironmentSatisfaction', 'HourlyRate',                     
                       'JobInvolvement', 'JobLevel','JobSatisfaction', 
                       'RelationshipSatisfaction', 'StockOptionLevel',
                        'TrainingTimesLastYear']].copy()
#print initial values
df_hr_data.head()
```

```python
#Copy categorical data
df_hr_categorical = df_hr[['Attrition', 'BusinessTravel','Department',
                       'EducationField','Gender','JobRole','MaritalStatus',
                       'Over18', 'OverTime']].copy()
#print initial values for cateforical
df_hr_categorical.head()
```

```python
df_hr_categorical.tail()
```

```python
df_hr_categorical.size
```

```python
#replace cateogrical values like yes or no to numberical values eg yes-1 /No-0 for attrition columns
Numeric_value = {'Yes':1, 'No':0}
df_hr_categorical['Attrition'] = df_hr_categorical["Attrition"].apply(lambda x: Numeric_value[x])
df_hr_categorical.head()
```

```python
#replacecategorical variables with dummy values
df_hr_categorical = pd.get_dummies(df_hr_categorical)
df_hr_categorical.head()
```

```python
df_hr_categorical.tail()
```

```python
df_hr_categorical.shape
```

```python
#final data numerical + categorical merge them
df_hr_data_final = pd.concat([df_hr_num, df_hr_categorical], axis=1)
df_hr_data_final.head()
```

```python
df_hr_data_final.tail()
```

```python
#modelling 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

target = df_hr_data_final['Attrition']
features = df_hr_data_final.drop('Attrition', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.4, random_state=10)
```

```python
#using RFC MODELL and fitting
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
#predicting test value 
test_pred = model.predict(X_test)
accuracy_score(y_test, test_pred)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=features.columns)
feat_importances = feat_importances.nlargest(20)
feat_importances.plot(kind='barh')
```
