```python
#feature engineering 
#plotting relevant graph for visulaization
def dist(df0,df1,df2,label0,label1,label2,features):
    plt.figure()
    fig,ax=plt.subplots(13,1,figsize=(8,45))
    i=0
    for ft in features:
        i+=1
        plt.subplot(13,1,i)
        
        sns.distplot(df0[ft], hist=False,label=label0)
        sns.distplot(df1[ft], hist=False,label=label1)
        sns.distplot(df2[ft], hist=False,label=label2)
        plt.xlabel(ft, fontsize=11)
        
        plt.tick_params(axis='x', labelsize=9)
        plt.tick_params(axis='y', labelsize=9)
    plt.show()
t0 = df_Skyserver.loc[df_Skyserver['class'] == 0]
t1 = df_Skyserver.loc[df_Skyserver['class'] == 1]
t2 = df_Skyserver.loc[df_Skyserver['class'] == 2]
features = df_Skyserver.columns.values[:13]
dist(t0,t1,t2, 'STAR', 'GALAXY','QSO', features)
```

```python
#from the output of the above.. taking into the consideration of redshift
df_Skyserver[['redshift','class']].groupby(['class'],as_index=False).mean().sort_values(by='class',ascending=False)
```

```python
#preparing data and modelling
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X=df_Skyserver.drop(['class'],axis=1)
y=df_Skyserver['class']
```

```python
X=StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=44, shuffle =True)
X_train
```

```python
#predicting using Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
random_forest = RandomForestClassifier(n_estimators=200)
random_forest.fit(X_train, y_train)
```

```python
y_pred = random_forest.predict(X_test)
Test_acc_random_forest = round(random_forest.score(X_test, y_test) * 100, 3)
print("Score from the prediction : ",Test_acc_random_forest)
```

```python
sns.heatmap(confusion_matrix(y_test , y_pred), center=True,annot=True,fmt='.1f')
```
